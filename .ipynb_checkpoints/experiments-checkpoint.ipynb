```python
# Load the data into variables for experimentation
job_description_path = "api/experiments/job_description.txt"
pdf_text_path = "api/experiments/pdf_text.txt"

try:
    with open(job_description_path, 'r', encoding="utf-8") as f:
        job_text = f.read()

    with open(pdf_text_path, 'r', encoding="utf-8") as f:
        cv_text = f.read()
        
    print("✅ Files loaded successfully!")
    print(f"Job Description length: {len(job_text)} characters")
    print(f"CV Text length: {len(cv_text)} characters")

except FileNotFoundError:
    print("❌ Error: Files not found. Make sure you ran the Django view first!")
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
                               
```

```python
df = pd.DataFrame([[job_text, cv_text]], columns=['job_text', 'cv_text'])
df['job_text_no_stopwords'] = df['job_text'].apply(lambda string: " ".join([word.lower() for word in string.split() if word not in en_stopwords]))
df['cv_text_no_stopwords'] = df['cv_text'].apply(lambda string: " ".join([word.lower() for word in string.split() if word not in en_stopwords]))
df.head()
```

```python
punct_pattern_to_remove = r'[^\w\s]'
df['job_text_no_punct'] = df['job_text_no_stopwords'].str.replace(punct_pattern_to_remove, "", regex=True)
df['cv_text_no_punct'] = df['cv_text_no_stopwords'].str.replace(punct_pattern_to_remove, "", regex=True)
df.head()
```

```python
df['tokenized_job'] = df.apply(lambda x: word_tokenize(x['job_text_no_punct']), axis = 1)
df['tokenized_cv'] = df.apply(lambda x: word_tokenize(x['cv_text_no_punct']), axis = 1)
df['tokenized_job'][0]
```

```python
ps = PorterStemmer()
df['stemmed_job'] = df['tokenized_job'].apply(lambda tokens: [ps.stem(token) for token in tokens])
df['stemmed_cv'] = df['tokenized_cv'].apply(lambda tokens: [ps.stem(token) for token in tokens])
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
df['lemmatized_job'] = df['tokenized_job'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
df['lemmatized_cv'] = df['tokenized_cv'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
df.head()
```

```python
job_tokens_clean =  sum(df['lemmatized_job'], [])
cv_tokens_clean =  sum(df['lemmatized_cv'], [])
```

```python

job_unigrams = pd.Series(nltk.ngrams(job_tokens_clean, 1)).value_counts()
job_unigrams

```

```python
cv_unigrams = pd.Series(nltk.ngrams(cv_tokens_clean, 1)).value_counts()
cv_unigrams
```

```python
job_bigrams = pd.Series(nltk.ngrams(job_tokens_clean, 2)).value_counts()
job_bigrams
```

```python
cv_bigrams = pd.Series(nltk.ngrams(cv_tokens_clean, 2)).value_counts()
cv_bigrams
```

```python
job_ngrams = pd.Series(nltk.ngrams(job_tokens_clean, 3)).value_counts()
job_ngrams
```

```python
cv_ngrams = pd.Series(nltk.ngrams(cv_tokens_clean, 3)).value_counts()
cv_ngrams
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
job_tokens_string = " ".join(job_tokens_clean)
cv_tokens_string = " ".join(cv_tokens_clean)

job_spacy_doc = nlp(job_tokens_string)
cv_spacy_doc = nlp(cv_tokens_string)

job_pos_df = pd.DataFrame(columns = ['token', 'pos_tag'])
cv_pos_df= pd.DataFrame(columns = ['token', 'pos_tag'])
```

```python
print(job_spacy_doc)
```

```python
print(cv_spacy_doc)
```

```python
job_pos_df.head()
```

```python
for token in job_spacy_doc:
    job_pos_df = pd.concat([job_pos_df,
                           pd.DataFrame.from_records([{'token': token.text, 'pos_tag':token.pos_}])], ignore_index = True)
```

```python
for token in cv_spacy_doc:
    cv_pos_df = pd.concat([cv_pos_df,
                           pd.DataFrame.from_records([{'token': token.text, 'pos_tag':token.pos_}])], ignore_index = True)
```

```python
job_pos_df.head(15)
```

```python
job_pos_df_count = job_pos_df.groupby(['token', 'pos_tag']).size().reset_index(name="counts").sort_values(by="counts", ascending= False)
cv_pos_df_count = cv_pos_df.groupby(['token', 'pos_tag']).size().reset_index(name="counts").sort_values(by="counts", ascending= False)
job_pos_df_count.head(15)
```
